[Quiz] iris 데이터에 대해 pairplot을 그려보자.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
from sklearn.datasets import load_iris
data = load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)
iris['Species'] = data.target
iris.Species = iris.Species.map(lambda x: data.target_names[x])

sns.pairplot(iris, hue='Species')
```

[Quiz]iris 데이터에 적용하고 hue를 Species로 지정해보자.

```python
plt.close('all')
pair_grid = sns.PairGrid(iris, hue='Species')
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.histplot) 
plt.show()
```

[Quiz] histplot에는 hue를 지정하지 않고자 한다.

```python
plt.close('all')
pair_grid = sns.PairGrid(iris, hue='Species')
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.histplot, hue=None) 
plt.show()
```
